```python
#Mengimport package yang digunakan untuk mengambil data dengan cara web scraping
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#Membuat request ke website
website='https://www.worldometers.info/coronavirus/#countries' # url for the site 
website_url=requests.get(website).text
soup = BeautifulSoup(website_url,'html.parser')
```

```python
#Mencari data dengan bagian tbody
my_table = soup.find('tbody')
```

```python
#Mengambil data
table_data = []
for row in my_table.findAll('tr'):
    row_data = []
    for cell in row.findAll('td'):
        row_data.append(cell.text)
    if(len(row_data) > 0):
        data_item = {"Country": row_data[1],
                     "TotalCases": row_data[2],
                     "NewCases": row_data[3],
                     "TotalDeaths": row_data[4],
                     "NewDeaths": row_data[5],
                     "TotalRecovered": row_data[6],
                     "ActiveCases": row_data[8],
                     "CriticalCases": row_data[9],
                     "Totcase1M": row_data[10],
                     "Totdeath1M": row_data[11],
                     "TotalTests": row_data[12],
                     "Tottest1M": row_data[13],
        }
        table_data.append(data_item)
```

```python
#Memasukan data kedalam dataframe
df = pd.DataFrame(table_data)
df.head(10)
```

```python
#Menghapus \n pada kolom Country
df['Country']=df['Country'].map(lambda x: x.lstrip('\n').rstrip('\n'))
df
```

```python
#Mengubah data kedalam bentuk excel, digunakan data hari Kamis, 8 April 2021 pukul 20.26 WIB
df.to_excel('Covid19_data.xlsx', index=True)
```
